```python
import numpy as np
import math
import cv2 as cv
import matplotlib.pyplot as plt

def croppingImg(image, center_point,size):
    
    ptX,ptY = center_point
    sizeX,sizeY = size
    
    img = image[ptX-(sizeX%2):ptX-(sizeX%2-sizeX),ptY-(sizeY%2):ptY-(sizeY%2-sizeY)]
    
    return img

def rotateImg(image,angle):
    rows,cols = image.shape
    M = cv.getRotationMatrix2D((cols/2,rows/2),angle,1) #(rotating_center,angle,scale)

    r = np.deg2rad(angle)
    new_rows, new_cols = (int(abs(rows*np.cos(r))+abs(cols*np.sin(r))),int(abs(cols*np.cos(r))+abs(rows*np.sin(r))))

    M[0,2] += (new_cols-cols)/2
    M[1,2] += (new_rows-rows)/2
    dst = cv.warpAffine(image,M,(new_cols,new_rows)) #(inputArray_src,rotationMat,size)

    return dst
```

```python
#reading image

img = cv.imread('image.jpeg',0)
```

```python
#cropping image

pt = tuple(int(x.strip()) for x in (input("center point :").split(',')))
size = tuple(int(x.strip()) for x in (input("size : ").split(',')))
    
crp = croppingImg(img,pt,size)
print(crp.shape)

plt.imshow(crp, cmap='gray')
plt.show()
```

```python
#rotating image

img = cv.imread('image.jpeg',0)

angle = input('Enter the angle of rotation : ')
    
dst = rotateImg(img,float(angle))

plt.imshow(dst, cmap='gray')
plt.show()
```
